```python
df=pd.get_dummies(df,columns=['Gender'])
```

```python
features=df.drop(['Gender_M','Gender_F','Orders'],axis=1)
features
```

```python
features.shape
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler().fit_transform(features)
```

```python
df_scale=pd.DataFrame(ss,columns=df.columns[1:36])
```

```python
df_scale
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

#list of diff value of k

n_cluster=[4,5,6,7,8]

           
# use of for loop 
           
           
for K in n_cluster:
    cluster=KMeans(n_clusters=K,random_state=10)
    predict=cluster.fit_predict(df_scale)
    score=silhouette_score(df_scale,predict,random_state=10)
    print(f'for n_clusters {K} , silhouette_score is {score}')
```

```python
model=KMeans(n_clusters=4,random_state=10)
model.fit(df_scale)
```

```python
df_cluster=df.copy(deep=True)
df_cluster['Cluster']=model.labels_
```

```python
df_cluster
```

```python
df_cluster['Cluster'].unique()
```

```python
#count of each cluster
np.unique(df_cluster['Cluster'],return_counts=True)
```

```python
sns.countplot(data=df_cluster,x='Cluster')
plt.title('cluster count',fontsize=12)

plt.text(x=-0.18,y=3000,s=np.unique(df_cluster['Cluster'],return_counts=True)[1][0])
plt.text(x=0.84,y=3000,s=np.unique(df_cluster['Cluster'],return_counts=True)[1][1])
plt.text(x=1.85,y=3000,s=np.unique(df_cluster['Cluster'],return_counts=True)[1][2])
plt.text(x=2.93,y=3000,s=np.unique(df_cluster['Cluster'],return_counts=True)[1][3])
```

```python
#cluster centers
df_center=pd.DataFrame(model.cluster_centers_,columns=df_cluster.columns[1:36])
```
